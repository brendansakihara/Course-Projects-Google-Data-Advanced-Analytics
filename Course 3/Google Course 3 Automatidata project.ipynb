```python
# Examine the unique values in the `passenger_count` column.
df['passenger_count'].value_counts()
```

```python
# Mean tips by passenger_count
tips_passenger_count_grouped = df.groupby(['passenger_count'])
tips_mean_passenger_count = tips_passenger_count_grouped['tip_amount'].mean()
tips_mean_passenger_count
```

```python
# Bar plot for mean tips by passenger count
plt.figure(figsize = (10,5))
sns.barplot(data = tips_mean_passenger_count.reset_index(),
            x = 'passenger_count',
            y = 'tip_amount');

plt.xlabel('Passenger Count')
plt.ylabel('Average Tip Amount')
plt.title('Aver Tip Amount by Passenger Count')
plt.show()
```

```python
# Create a month column
df['month'] = df['tpep_pickup_datetime'].dt.month_name()
# Create a day column
df['day'] = df['tpep_pickup_datetime'].dt.day_name()

df.head()
```

```python
# Get total number of rides for each month
month_ride_counts = df['month'].value_counts()
month_ride_counts
```

```python
# Reorder the monthly ride list so months go in order
month_list = ['January', 'February', 'March', 'April', 'May', 'June', 'July','August', 'September', 'October', 'November', 'December']

month_ride_counts = month_ride_counts.reindex(index = month_list)
month_ride_counts
```

```python
month_ride_counts.index
```

```python
# Bar plot of total rides per month
plt.figure(figsize = (10,5))
sns.barplot(x = month_list, y = month_ride_counts)
plt.xlabel('Month')
plt.ylabel('Ride Counts')
plt.title('Ride Counts by Month')
plt.xticks(rotation = 45)
plt.show()
```

```python
# Repeat the above process, this time for rides by day
day_list = ['Sunday', 'Monday', 'Tuesday', 'Wednesday', 'Thursday', 'Friday', 'Saturday']

day_rides_count = df['day'].value_counts()
day_rides_count = day_rides_count.reindex(index = day_list)
day_rides_count
```

```python
plt.figure(figsize = (10,5))
sns.barplot(x = day_list, y = day_rides_count)
plt.xlabel('Weekday')
plt.ylabel('Count')
plt.title('Ride Counts by Weekday')
plt.xticks(rotation = 45)
plt.show()
```

```python
# Repeat the process, this time for total revenue by day
day_grouped = df.groupby(['day'])
day_total_revenue = day_grouped['total_amount'].sum()
day_total_revenue = day_total_revenue.reindex(index = day_list)
day_total_revenue
```

```python
# Bar plot of total revenue by day
plt.figure(figsize = (10,5))
sns.barplot(x = day_list, y = day_total_revenue)
plt.xlabel('Weekday')
plt.ylabel('Total Revenue')
plt.title('Total Revenue for Weekday')
plt.show()
```

```python
# Repeat the process, this time for total revenue by month
month_grouped = df.groupby(['month'])
month_total_revenue = month_grouped['total_amount'].sum()
month_total_revenue = month_total_revenue.reindex(index = month_list)
month_total_revenue
```

```python
# Bar plot of total revenue by month
plt.figure(figsize = (10,5))
sns.barplot(x = month_list, y = month_total_revenue)
plt.xlabel('Month')
plt.ylabel('Total Revenue')
plt.title('Total Revenue for Month')
plt.show()
```
